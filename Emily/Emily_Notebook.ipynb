```python
import pandas as pd
```

```python
df = pd.read_csv('Data/AviationData.csv', encoding='cp1252')
```

Cleaning Data
1. Cleaning columns to uniform formatting
2. making Location and Make uniform formatting
3. Converting Event Date to Date Time format
4. creating Make and Model column
5. creating year and month rows
6. Sort by Event Date

```python
df.columns = df.columns.str.replace(' ', '')
df.columns = df.columns.str.replace('.', ' ')
df.columns = df.columns.str.title()
df['Location'] = df['Location'].str.title()
df['Make'] = df['Make'].str.title()
df['Year'] = [int(x[:4]) for x in df['Event Date']]
df['Month'] = [int(x[5:7]) for x in df['Event Date']]
df['Make & Model'] = df['Make'] + " " + df['Model']
df = df[df['Year'] >= 2003].copy()
```

```python
df = df[df['Country'] == 'United States'].copy()
```

```python
df['State'] = [str(x[-2:]) for x in df['Location']]
df['State'] = df['State'].str.upper()
df.head()
```

Handling NaNs:
1. remove Schedule column
2. remove Broad Phase of Flight column
3. remove rows with Nan in Location
4. remove rows with nan for total fatal injuries, total minor, total uninjured


```python
df.drop(['Schedule', 'Broad Phase Of Flight'], axis=1, inplace=True)
df.dropna(subset = ['Location', 'Total Fatal Injuries', 'Total Serious Injuries', 'Total Minor Injuries', 'Total Uninjured'], inplace=True)
df['Total Passengers'] = df['Total Fatal Injuries'] + df['Total Serious Injuries'] + df['Total Minor Injuries'] + df['Total Uninjured']
df.info()
```

```python
df['Total Passengers'] = df['Total Fatal Injuries'] + df['Total Serious Injuries'] + df['Total Minor Injuries'] + df['Total Uninjured']
df['Percent Fatal Injuries'] = (df['Total Fatal Injuries'] / df['Total Passengers'])*100
df['Percent Serious Injuries'] = (df['Total Serious Injuries'] / df['Total Passengers'])*100
df['Percent Minor Injuries'] = (df['Total Minor Injuries'] / df['Total Passengers'])*100
df['Percent Uninjured'] = (df['Total Uninjured'] / df['Total Passengers'])*100
df['Percent Fatal Injuries'] = df['Percent Fatal Injuries'].round(4)
df['Percent Serious Injuries'] = df['Percent Serious Injuries'].round(4)
df['Percent Minor Injuries'] = df['Percent Minor Injuries'].round(4)
df['Percent Uninjured'] = df['Percent Uninjured'].round(4)
```
